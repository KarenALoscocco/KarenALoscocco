I used the following reference for the idea to model this problem as a shortest path problem: https://www.lix.polytechnique.fr/~liberti/teaching/isic/isc612-06/exercises.pdf   

The possibilities for this renewal plan can be modeled by the following graph: 

![alt text](1diagram.jpg "Title")

The nodes 1 to 5 represent the start of each year. The 6th node represents the end of the 5th year or end of the planning period. For each node $i \lt 6$ and $i \lt j$, arc $(i,j)$ represents the event that machinery bought at the beginning of the i-th year is sold at the beginning of the j-th year. The cost (in thousands of dollars) $c_{ij}$ along each arc $(i,j)$ is given by:   
$$ c_{ij} = 12 + \big( \sum_{k = i}^{j-1} m_k \big) - g_j$$   
where $m_k$ is the maintenance cost in the k-th year and $g_j$ is the gain from selling the machinery at the beginning of the j-th year.     
&nbsp;   
A renewal plan which minimizes the total operation cost is the shortest path from node 1 to 6 becasue the cost of the path is the cost of the plan.

The following is the integer program:  

$\text{Nodes } V = \{1,2,3,4,5,6\} \\   
\text{Arcs }A = \{(1,2),(1,3),(1,4),(1,5),(1,6),(2,3),(2,4),(2,5),(2,6),(3,4),(3,5),(3,6),(4,5),(4,6),(5,6) \}$ 

$x_{ij} = \bigg\{
\begin{align} 
1 && \text{ if } i \rightarrow j \text{ is a path } p \\
0 && \text{otherwise} \end{align} $ 

$$\text{min } \sum_{(i,j) \in A} c_{ij}x_{ij} \\   
\sum_{i: (1,i) \in A } x_{1,i} = 1 \\    
\sum_{i: (i,6) \in A} x_{i,6} = 1 \\    
\forall j \in V \backslash \{1,6\} \sum_{i:(i,j) \in A} x_{ij}-\sum_{i:(j,i) \in A}x_{ji} = 0   \\   
x_{ij} \in \{0,1\} \forall i,j$$   
&nbsp;   
The following is the solution in Gurobi:   

```python
from gurobipy import *
import pandas as pd
import numpy as np
from itertools import permutations
from itertools import product 
import matplotlib.pyplot as plt
from sympy import *

def printOptimal(m):
    print('\n')
    if m.Status == GRB.OPTIMAL:
        print('Variable Values:')
        for v in m.getVars():
            print(v.VarName, v.X)
        
        print('\nObjective Value: {}\n'.format(str(m.objVal)))
```

```python
V = set([1,2,3,4,5,6])
c = np.array([7,12,21,31,44,7,12,21,31,7,12,21,7,12,7])
m = Model()
X = m.addVars([i for i in permutations(V,2) if i[1]>i[0]],vtype = GRB.BINARY, name = 'X')
m.addConstr(sum(X[1,i] for i in V if i!=1) == 1)
m.addConstr(sum(X[i,6] for i in V if i!=6) == 1)

j = V-{1,6}

for v in j: 
    m.addConstr(sum(X[i,v] for i in V if i<v) - sum(X[v,i] for i in V if i>v) == 0)
    
m.setObjective(np.dot(c,X.values()), GRB.MINIMIZE)

m.optimize()

printOptimal(m)
```

The following is the solution in Gurobi:   

```python
V = set([1,2,3,4,5,6])
c = np.array([7,12,21,31,44,7,12,21,31,7,12,21,7,12,7])
m = Model()
X = m.addVars([i for i in permutations(V,2) if i[1]>i[0]], name = 'X')
m.addConstr(sum(X[1,i] for i in V if i!=1) == 1)
m.addConstr(sum(X[i,6] for i in V if i!=6) == 1)

j = V-{1,6}

for v in j: 
    m.addConstr(sum(X[i,v] for i in V if i<v) - sum(X[v,i] for i in V if i>v) == 0)
    
m.setObjective(np.dot(c,X.values()), GRB.MINIMIZE)

m.optimize()

printOptimal(m)
```

The linear program relaxation will always give the integer solution for free.

## 3 (b)    
The following table outlines the penalties for the i-th unit of increase in B:   

|$i$| penalty $p_i$ for the $i$-th unit|
|----|---|
| 1| 2|
| 2| 3| 
| 3| 3| 
| 4| 6| 
| 5| 6| 
| 6|10| 
| 7|10|
| 8|10|
| 9|10|
|10|10|

B can be increased by a certain amount, $S = \sum_i a_ix_i - B$   

$y_{i} = \bigg\{
\begin{align} 
1 && \text{ if } S \text{ is } \geq i && \forall i = 1,\dots, 10\\
0 && \text{otherwise} \end{align}\\  $    

For example, if B increases by 6, then $y_1 = y_2 = y_3 = y_4 = y_5 = y_6 = 1$ and then this will increase costs by $1(2)+1(3)+1(3)+1(6)+1(6)+1(10) = 30$.   

The cost function is altered to include this added penalty cost ($\sum p_iy_i$).   

The following logic can be used to create the constraints:    
If $S-1\geq 0$ then $y_1=1$   
If $S-2\geq 0$ then $y_2=1$   
If $S-3\geq 0$ then $y_3=1$   
And so on...    
The constraints become:    
$-(y_1-1) \leq Mz_1 \;,\; S-1 \leq M(1-z_1)$   
$-(y_2-1) \leq Mz_2 \;,\; S-2 \leq M(1-z_2)$   
$-(y_3-1) \leq Mz_3 \;,\; S-3 \leq M(1-z_3)$   

The following is the integer linear program:    

$$\text{min } \sum_jc_jx_j + \sum_ip_iy_i\\    
\sum_j a_jx_j - B \leq S \\   
-(y_i-1) \leq Mz_i \; \forall i = 1,\dots, 10 \\   
S-i \leq M(1-z_i) \; \forall i = 1,\dots, 10\\   
x_j \in \{0,1 \} \\   
y_i \in \{0,1 \} \\   
S \geq 0\\   
z_i \in \{0,1\}$$

```python
p = np.array([2,3,3,6,6,10,10,10,10,10])  
n = 10
B = 30
a = np.ceil(np.random.rand(10)*10)
c = np.ceil(np.random.rand(10)*10)
M = 1000

m = Model()
X = m.addVars(n,vtype = GRB.BINARY, name = 'X')
Y = m.addVars(n,vtype = GRB.BINARY, name = 'Y')
s = m.addVar(name = 's',ub = 10)
z = m.addVars(n,vtype = GRB.BINARY, name = 'z')

m.addConstr(s == np.dot(a,X.values())-B)

for i in range(9): 
    m.addConstr(-Y[i]+1 <= M*z[i])
    m.addConstr(s-i <= M*(1-z[i]))

obj = np.dot(c,X.values()) - np.dot(p,Y.values())

m.setObjective(obj, GRB.MAXIMIZE)

m.optimize()

printOptimal(m)
```

## 4 (a)

$w_j\text{: number of of cores per program}$   
$q_j\text{: number of programs desired}$   
$x_{ij}\text{: number of programs with } w_j\text{ number of cores}$

Let $F = \{ P = (a_{P,1},\dots,a_{P,n}): \sum_{i=1}^n w_ja_{P,j} \leq W , a_{P,j} \in Z_{+} \; \forall 1\leq j \leq n\}$ be the set of all valid patterns. 

Then solve:    
$$\text{min } \sum_{P \in F}x_P \\   
\sum_{P\in F}x_p a_{P,j} = b_j \; \forall j = 1,\dots,n \\   
x_p \geq 0 \; \forall P \in F$$


```python
w = np.array([20,30,32,16,5])
q = np.array([1000,1500,2000,2500,1200])
```

Get all possible patterns:   

```python
L = min(w)
R = 64
n = len(w)

m = Model()

m.params.PoolSolutions = 400

m.params.PoolSearchMode = 2

y = m.addVars(n,vtype = GRB.INTEGER, name = 'Y')

m.addConstr(np.dot(w,y.values()) <= R)

m.setObjective(0, GRB.MINIMIZE)

m.optimize()

printOptimal(m)
```

```python
NumSolns = m.solCount

solutions = np.empty([n, NumSolns])

for x in range(NumSolns):
    m.params.outputFlag = 0
    m.params.SolutionNumber = x
    for i in range(n):
        solutions[i,x] = m.Xn[i]
```

There are {{NumSolns}} possible patterns.    

Now solve the LP:   

```python
def printOptimal(m):
    print('\n')
    if m.Status == GRB.OPTIMAL:
        
        print('\nObjective Value: {}\n'.format(str(m.objVal)))
```

```python
m = Model()

x = m.addVars(NumSolns, vtype = GRB.CONTINUOUS, name = 'X')
        
for i in range(solutions.shape[0]):
    m.addConstr(np.dot(solutions[i,:],x.values()) >= q[i])
                
m.setObjective(sum(x.values()), GRB.MINIMIZE)

m.optimize()
                
printOptimal(m)
```

Now solved with Column Generation:   

```python
def printOptimal(m):
    print('\n')
    if m.Status == GRB.OPTIMAL:
        print('Variable Values:')
        for v in m.getVars():
            print(v.VarName, v.X)
        
        print('\nObjective Value: {}\n'.format(str(m.objVal)))
```

```python
def solveRestrictedLP(I):

    m = Model()
    
    m.params.outputFlag = 0

    x = m.addVars(len(I), vtype = GRB.CONTINUOUS, name = 'X')

    for i in range(solutions.shape[0]):
        m.addConstr(np.dot(solutions[i,I],x.values()) >= q[i])

    m.setObjective(sum(x.values()), GRB.MINIMIZE)

    m.optimize()
    
    y = m.Pi
    
    return(m.X, m.VBasis, m.objVal, y)
```

```python
def solveReducedCostLP(y):
    m = Model()
    
    m.params.outputFlag = 0
    
    a = m.addVars(solutions.shape[0], vtype = GRB.INTEGER, name = 'A')
    
    m.addConstr(np.dot(w,a.values()) <= 64)
    
    m.setObjective((1 - np.dot(a.values(),y)) , GRB.MINIMIZE)
    
    m.optimize()
    
    for i in range(solutions.shape[1]):
        if (list(solutions.T[i]) == list(np.floor(m.X))):
            newpattern = i
            
    return(m.X, m.objVal, newpattern)
```

```python
w = np.array([20,30,32,16,5])
q = np.array([1000,1500,2000,2500,1200])

I = []

for i in range(NumSolns):
    if sum(solutions[:,i]) == 1: 
        I.append(i)
```

```python
obj = -10000
while obj < 0: 
    X, VBasis, objVal, y = solveRestrictedLP(I)
    x, obj, newpattern = solveReducedCostLP(y)
    I.append(newpattern)
    print('-----')
    print(obj)
    print(objVal)
```
